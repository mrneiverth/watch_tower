```python
import pandas as pd 
import numpy as np
```

```python
path = 'steam-200k.csv'
#path = 'steam-200k.csv'
df = pd.read_csv(path, names = ['UserID', 'Game', 'Action', 'Hours', 'Not Needed'])
df.head()
```

```python
# Creating a new variable 'Hours Played' and code it as previously described.
df['Hours_Played'] = df['Hours'].astype('float32')

df.loc[(df['Action'] == 'purchase') & (df['Hours'] == 1.0), 'Hours_Played'] = 0

# Sort the df by User ID, games, and hours played
# Drop the duplicated records, and unnecessary columns
df.UserID = df.UserID.astype('int')
df = df.sort_values(['UserID', 'Game', 'Hours_Played'])

clean_df = df.drop_duplicates(['UserID', 'Game'], keep = 'last').drop(['Action', 'Hours', 'Not Needed'], axis = 1)

# every transaction is represented by only one record now
clean_df.head()
```

```python
#Salva o dataframe no csv
clean_df.to_csv('Games_With_Hours.csv', index = False)

#pega apenas a coluna de jogos e elimina os duplicados
data = clean_df.drop(['UserID', 'Hours_Played'], axis=1)
data = data.drop_duplicates('Game', keep = 'last')

#Organiza os jogos em ordem alfabetica
data = data.sort_values('Game')

#Salva em um CSV
data.to_csv('game_Names.csv', index = False)
data.head()
```
